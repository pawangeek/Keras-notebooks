```python
import numpy as np
from keras.datasets import mnist
from keras.optimizers import SGD
from keras.models import Sequential
from keras.layers.core import Dense,Activation,Dropout

from keras.utils import np_utils
```

```python
epoch = 200
batch_size = 128
verbose = 1
classes = 10 #no. of outputs
optimizer = SGD() 
hidden = 128
val_split = 0.2 #data reserved for validation
```

```python
(X_train,y_train), (X_test, y_test) = mnist.load_data()
#data shuffled n=between train and test sets
```

```python
X_train.shape
```

```python
reshape = 784

X_train = X_train.reshape(60000,reshape)
X_test = X_test.reshape(10000,reshape)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
X_train /= 255
X_test /= 255

print(X_train.shape[0])
print(X_test.shape[0])
```

```python
y_train
```

```python
Y_train = np_utils.to_categorical(y_train,classes)
Y_test = np_utils.to_categorical(y_test,classes)
```

```python
#input layer = 28*28 = 784 neurons, one for each pixel in images
# Further values associated with each pizel are normalized in range[0,1]
```

```python
model = Sequential()
model.add(Dense(classes,input_shape=(reshape,)))
model.add(Activation('softmax'))
```

Softmax is generalization of sigmoid function
<br> Here, it aggregates 10 answers provided by previous layer with 10 neurons

```python
model.summary()
# param = (784+1(bias))x10 = 7850
```

```python
model.compile(loss='categorical_crossentropy', optimizer=optimizer,metrics =['accuracy'])
```

```python
history=model.fit(X_train,Y_train,batch_size=batch_size, epochs=epoch,verbose=verbose,validation_split=val_split)
```

```python
score = model.evaluate(X_test, Y_test, verbose=verbose)
```

```python
score
```

congratulations, you have just defined your first neural network in Keras. A few lines of
code, and your computer is able to recognize handwritten numbers. 

## Adding more hidden layers with ReLU

```python
model = Sequential()
model.add(Dense(hidden,input_shape = (reshape,) ))
model.add(Activation ('relu'))
model.add(Dense(hidden))
model.add(Activation ('relu'))
model.add(Dense(classes))
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer=optimizer,metrics=['accuracy'])
```

```python
history = model.fit(X_train,Y_train,batch_size=batch_size,epochs=epoch, verbose=verbose,validation_split=val_split)
```

```python
score = model.evaluate(X_test, Y_test, verbose=verbose)
```

```python
score
```

## Adding Dropout to layers

 We decide to randomly drop with the dropout
probability some of the values propagated inside our internal dense network of hidden
layers. In machine learning, this is a well-known form of regularization

```python
model = Sequential()
model.add(Dense(hidden,input_shape=(reshape,)))
model.add(Activation('relu'))
model.add(Dropout(0.3))
model.add(Dense(hidden))
model.add(Activation('relu'))
model.add(Dropout(0.3))
model.add(Dense(classes))
model.add(Activation('softmax'))
```

## What is dropout

The idea of droupout is simple: During the forward pass of your network just randomly set some of the neurons to zero.
<p>With dropout, the weights of the nodes learned through backpropagation become somewhat more insensitive to the weights of the other nodes and learn to decide the outcome independent of the other neurons.<p>Hence, every neuron learns how to take individuals responsibility and performance, as no other neurons are available permanently to rely on.<p>But finally, at the time of test, all neurons participate with proper fraction, like an ensemble of networks, which improves the performance of network in a mechanism called regularizatio

```python
history = model.fit(X_train,Y_train,batch_size=batch_size,epochs=epoch, verbose=verbose,validation_split=val_split)
```

```python
score = model.evaluate(X_test, Y_test, verbose=verbose)
```
